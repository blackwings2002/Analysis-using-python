# Covid-19 Analysis Project 


```python
#importing libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#import covid dataset 
ds = pd.read_csv('covid19_Confirmed_dataset.csv')
ds.head()
ds.drop(['Lat','Long'],axis=1,inplace=True) #axis 1 = columns
```

```python
ds.head()
```

```python
#aggregating the coutnry
ds_aggregated= ds.groupby('Country/Region').sum()
ds_aggregated
```

```python
ds_aggregated.drop('Province/State',axis=1,inplace=True)
ds_aggregated
```

```python
ds_aggregated.shape
```

```python
#Visualise the country data

ds_aggregated.loc['China'].plot()
ds_aggregated.loc['India'].plot()
ds_aggregated.loc['Spain'].plot()
plt.legend()
plt.show()
```

```python
# for first 3 days in china 
ds_aggregated.loc['China'][:3].plot()
```

```python
#find the first derivative of the curve
ds_aggregated.loc['China'].diff().plot()
```

```python
#find the maximum infection rate for each country
ds_aggregated.loc['China'].diff().max()
```

```python
ds_aggregated.loc['India'].diff().max()
```

```python
ds_aggregated.loc['Spain'].diff().max()
```

```python
#finding maximum infection rate for each country
ds_aggregated_list=list(ds_aggregated.index)
max_infection_rate=[]

for i in ds_aggregated_list:
    a=ds_aggregated.loc[i].diff().max()
    max_infection_rate.append(a)
ds_aggregated['max_infection_rate'] = max_infection_rate
ds_aggregated
```

```python
#Create a new DataFrame
Corona_data =pd.DataFrame(ds_aggregated['max_infection_rate'])
Corona_data
```

```python
#importing dataset 2
happiness_report= pd.read_csv('worldwide_happiness_report.csv')
happiness_report.head()
```

```python
#clean the data by dropping useless column

useless_col=["Overall rank","Score","Generosity","Perceptions of corruption"]

happiness_report.drop(useless_col,axis=1,inplace=True)
happiness_report.set_index("Country or region")
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report
```

```python
#join the two dataset using join 
data=Corona_data.join(happiness_report,how='inner')
data
```

```python
data.corr()  
```

```python
#lets plot 
a = data['max_infection_rate']
b = data['GDP per capita']

sns.scatterplot(x=b,y=np.log(a)) #logscaling 
```

```python
#creating regression model
sns.regplot(x=b,y=np.log(a))
```

```python
#lets plot max_infection_rate and social support

x= data['max_infection_rate']
y= data['Social support']

sns.scatterplot(x=y,y=np.log(x)) #using logscaling to compress the wide-ranging values
```

```python
#creagting regression plot for this
sns.regplot(x=y,y=np.log(x))
```
